```python
# Преобразование типов
data = data.astype({
    "Transaction_ID": "Int64",
    "Customer_ID": "Int64",
    "Zipcode": "Int64",
    "Age": "Int64",
    "Ratings": "Int64",
    "Phone": "Int64",
    "Total_Purchases": "Int64",
    "Year": "Int64"
})

# Переменная с финальной суммой строки товара
data['Total_Price'] = data['Total_Purchases'] * data['Amount']
```

# Выручка

```python
# Группировка по годам
yearly_revenue = data.groupby('Year')['Total_Price'].sum().reset_index()
print(yearly_revenue)

# Группировка по годам и месяцам
monthly_revenue = data.groupby(['Year', 'Month'])['Total_Price'].sum().reset_index()
monthly_revenue['Month'] = pd.to_datetime(monthly_revenue['Month'], format='%B').dt.month

monthly_revenue= monthly_revenue.sort_values(by=['Year', 'Month'])
monthly_revenue
```

### Динамика выручки по месяцам

```python
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
import seaborn as sns

# Сортируем данные по году и месяцу
monthly_revenue = monthly_revenue.sort_values(by=['Year', 'Month'])

# Создаем новый столбец с номером месяца для оси X (например, 1 для Января, 2 для Февраля и т.д.)
monthly_revenue['Month_Year'] = monthly_revenue['Year'].astype(str) + '-' + monthly_revenue['Month'].astype(str)

# Построение графика
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_revenue, x='Month_Year', y='Total_Price', marker='o')

# Настроим подписи оси X с месяцы, которые будут повторяться для каждого года
months_labels = ["Январь", "Февраль", "Март", "Апрель", "Май", "Июнь", "Июль", "Август", "Сентябрь", "Октябрь", "Ноябрь", "Декабрь"]
plt.xticks(ticks=range(len(monthly_revenue)), labels=months_labels * len(monthly_revenue['Year'].unique()))

# Добавление вертикальных линий для разделения лет
for year in monthly_revenue['Year'].unique():
    year_months = monthly_revenue[monthly_revenue['Year'] == year]
    plt.axvline(x=year_months['Month'].iloc[-1], linestyle='--', color='black', alpha=0.5)


# Улучшаем ось Y, чтобы выручка отображалась более читаемо
def y_axis_formatter(x, pos):
    return f'{int(x):,}'  # Форматируем числа с разделением тысяч

plt.gca().yaxis.set_major_formatter(FuncFormatter(y_axis_formatter))


# Добавление заголовков и подписей
plt.title('Динамика выручки по месяцам и годам')
plt.xlabel('Месяц')
plt.ylabel('Выручка')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

# 3. Средний чек



## Средний чек по годам по всем товарам

```python
yearly_avg_check = data.groupby('Year')[['Total_Price', 'Total_Purchases']].sum()

# Рассчитываем средний чек
yearly_avg_check['Average_Check'] = yearly_avg_check['Total_Price'] / yearly_avg_check['Total_Purchases']
yearly_avg_check
```

## За 2024 по месяцам

```python
data_2024 = data[data['Year'] == 2024]

# Преобразуем столбец 'Month' в категориальный тип с порядком месяцев
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

data_2024['Month'] = pd.Categorical(data_2024['Month'], categories=month_order, ordered=True)

# Группируем по месяцам и вычисляем суммарные Total_Price и Total_Purchases
monthly_avg_check = data_2024.groupby('Month')[['Total_Price', 'Total_Purchases']].sum()

# Рассчитываем средний чек
monthly_avg_check['Average_Check'] = monthly_avg_check['Total_Price'] / monthly_avg_check['Total_Purchases']

# Посмотрим на результат
monthly_avg_check
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Строим график динамики среднего чека по месяцам
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_avg_check, x='Month', y='Average_Check', marker='o')

plt.title('Динамика среднего чека по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Средний чек')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# DAU / MAU

```python
from dateutil import parser

def parse_date(date_str):
    try:
        return parser.parse(date_str, dayfirst=False)  # Всегда сначала месяц
    except:
        return pd.NaT  # Если не удалось распарсить, вернуть NaT

data['Date'] = data['Date'].apply(parse_date)
```

```python
print('Количество пропущенных значений дат:', data['Date'].isna().sum())
```

## DAU
Daily Active Users


```python
december_2023 = data[(data['Year'] == 2023) & (data['Month'] == 'December')]
december_2023 = december_2023.groupby('Date').size().reset_index()
december_2023 = december_2023[december_2023['Date'] >= '2023-12-01'].reset_index()

december_2023 = december_2023.drop(columns=['index'])
december_2023.columns = ['Date', 'Unique_Users']

december_2023.sort_values(by='Unique_Users', ascending=False ).head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Строим график динамики уникальных посетителей по дням декабря
plt.figure(figsize=(8, 5))
sns.lineplot(data=december_2023, x='Date', y='Unique_Users', marker='o')

# Настроим отображение только дней декабря на оси X
plt.title('Динамика уникальных посетителей по дням декабря 2023')
plt.xlabel('День декабря')
plt.ylabel('Уникальные посетители')

# Настроим метки оси X так, чтобы показывались только числа
plt.xticks(ticks=pd.date_range('2023-12-01', '2023-12-31', freq='D'), labels=range(1, 32))

plt.grid(True)
plt.show()
```

## MAU
Monthly Active Users

```python
data_2023 = data[data['Year'] == 2023]

# Преобразуем столбец 'Month' в категориальный тип с порядком месяцев
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

data_2023['Month'] = pd.Categorical(data_2023['Month'], categories=month_order, ordered=True)


# Группируем по месяцу и считаем количество уникальных пользователей
mau = data_2023.groupby('Month')['Customer_ID'].nunique().reset_index()

# Переименовываем столбцы
mau.columns = ['Month', 'MAU']

mau.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Строим график динамики среднего чека по месяцам
plt.figure(figsize=(8, 5))
sns.lineplot(data=mau, x='Month', y='MAU', marker='o')

plt.title('Динамика уникальных посетителей по месяцам в 2023')
plt.xlabel('Месяц')
plt.ylabel('Уникальные посетители')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# Среднее колличесво позиций в чеке

```python
# Группируем по Transaction_ID и суммируем количество товаров в каждой транзакции
total_purchases_per_transaction = data.groupby('Transaction_ID')['Total_Purchases'].sum().reset_index()

# Рассчитываем среднее количество позиций на чек
average_positions_per_transaction = total_purchases_per_transaction['Total_Purchases'].mean()

# Выводим результат
print(f"Среднее количество позиций в чеке: {average_positions_per_transaction:.2f}")
```

```python
# Отфильтруем данные за 2024 год
data_2024 = data[data['Year'] == 2024]

# Группируем по месяцу и Transaction_ID, суммируем Total_Purchases по транзакциям
monthly_total_purchases = data_2024.groupby(['Month', 'Transaction_ID'])['Total_Purchases'].sum().reset_index()

# Рассчитываем среднее количество позиций в чеке по месяцам
monthly_avg_positions = monthly_total_purchases.groupby('Month')['Total_Purchases'].mean().reset_index()
monthly_avg_positions.sort_values(by='Total_Purchases')
#
```
